# Fire Disturbance Analysis in Western US Forests

This notebook processes forest disturbance data across the Western United States to identify areas impacted by wildfires over time. The workflow involves reading forest disturbance data for each year, isolating fire-related disturbances, and generating raster files that show fire presence across different time periods.

The analysis includes:
- Identifying fire-impacted areas by processing annual forest disturbance raster data.
- Aggregating fire presence data over five-year periods to track regions experiencing two or more years of fire-related forest disturbance.
- Summarizing the total area affected by fire over five-year intervals and visualizing these trends with a plot.

The output includes:
- Geospatial raster files showing fire presence per year.
- Five-year aggregated raster files showing the total number of fire-impacted years.
- A plot visualizing the total area affected by fires over five-year periods.

This analysis helps track the extent of wildfire disturbances in Western US forests over time.

```python
# Import packages
import os
from glob import glob
import earthpy as et
from os import path
import pandas as pd
import matplotlib.pyplot as plt
import rasterio as rio
import gc

home = path.join(et.io.HOME, "GitHub", "visualize-forest-disturbance")

forest_folder = path.join(home, "forest-disturbance-stack", "western-conus")
fire_folder = path.join(home, "forest-disturbance-stack", "fire-western-conus")
fiveyear_fire_folder = path.join(home, "forest-disturbance-stack", "five-year-fire")

forest_tif_list = sorted(glob(path.join(forest_folder, "forest-disturbance-stack_western-conus_*.tif")))
```

```python
# Loop through each year of forest disturbances for a 5 year period and calculate where drought was present
for tif in forest_tif_list:
    year = tif[-8:-4]
    print(year)
    with rio.open(tif) as src:
        array = src.read(1)
        meta = src.profile
    array[array == 1] = 1
    array[array == 5] = 1
    array[array == 9] = 1
    array[array == 13] = 1
    array[array != 1] = 0
    with rio.open(path.join(fire_folder, "fire_presence_{}.tif".format(year)), 'w', **meta) as dst:
        dst.write(array, 1)
```

```python
# Find all areas that had more than one drought year in each time period
fire_tif_list = sorted(glob(path.join(fire_folder, "fire_presence_*.tif")))
year_totals_over_2 = {}

for i, tif in enumerate(fire_tif_list):
    if i >=4:
        years = fire_tif_list[i-4:i+1]
        start_year = years[0][-8:-4]
        end_year = years[4][-8:-4]
        print("Processing years {} to {}".format(start_year, end_year))
        with rio.open(years[0]) as src:
            years_array = src.read(1)
            meta = src.profile
        for year in years[1:]:
            with rio.open(year) as src:
                array = src.read(1)
            years_array += array
            del(array)
            gc.collect()
        with rio.open(path.join(fiveyear_fire_folder, "fire_totals_{}-.{}.tif".format(start_year, end_year)), 'w', **meta) as dst:
            dst.write(years_array, 1)
        years_array[years_array <= 1] = 0
        years_array[years_array > 1] = 1
        year_totals_over_2[end_year] = (years_array.sum()* 30 * 30)/1000000    
        del(years_array)
        gc.collect()
```

```python
# Rearrange data and plot it
years = list(year_totals_over_2.keys())
area = list(year_totals_over_2.values())
data = {'End Year': years, 'Total Area': area}
short_term_plot = pd.DataFrame(data)

fig, ax = plt.subplots()
short_term_plot.plot('End Year',
                    y='Total Area', 
                    title='Total area that experienced 2+ fires within 5 years in Western US Forests',
                    ax=ax)
plt.savefig(os.path.join(fiveyear_fire_folder, "fiveyear_fire.png"))
plt.show()
```

```python
short_term_plot.to_csv(os.path.join(fiveyear_fire_folder, 'fire_area_over_time.csv'))
```
