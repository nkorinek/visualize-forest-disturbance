# Visualize Moving Windows Through Plots

A notebook to linearly visualize disturbance totals over time using the five year moving window

```python
# Import packages and define file paths
import os
import earthpy as et
from os import path
import pandas as pd
import matplotlib.pyplot as plt

home = path.join(et.io.HOME, "GitHub", "visualize-forest-disturbance")

forest_folder = path.join(home, "forest-disturbance-stack", "western-conus")
fire_folder = path.join(home, "forest-disturbance-stack", "fire-western-conus")
fiveyear_fire_folder = path.join(home, "forest-disturbance-stack", "five-year-fire")
combine_folder = path.join(home, "forest-disturbance-stack", "combined-western-conus")
beetle_folder = path.join(home, "forest-disturbance-stack", "beetle-western-conus")
fiveyear_beetle_folder = path.join(home, "forest-disturbance-stack", "five-year-beetle")
drought_folder = path.join(home, "forest-disturbance-stack", "drought-western-conus")
fiveyear_drought_folder = path.join(home, "forest-disturbance-stack", "five-year-drought")
fiveyear_fire_beetle_folder = path.join(home, "forest-disturbance-stack", "five-year-fire-beetle")
fiveyear_all_folder = path.join(home, "forest-disturbance-stack", "five-year-all")
all_folder = path.join(home, "forest-disturbance-stack", "all-western-conus")
```

```python
# Read CSV data
beetle_plot = pd.read_csv(os.path.join(fiveyear_beetle_folder, 'beetle_area_over_time.csv'), parse_dates=['End Year'])
fire_plot = pd.read_csv(os.path.join(fiveyear_fire_folder, 'fire_area_over_time.csv'), parse_dates=['End Year'])
drought_plot = pd.read_csv(os.path.join(fiveyear_drought_folder, 'drought_area_over_time.csv'), parse_dates=['End Year'])
fire_beetle_plot = pd.read_csv(os.path.join(fiveyear_fire_beetle_folder, 'fire_beetle_area_over_time.csv'), parse_dates=['End Year'])
all_plot = pd.read_csv(os.path.join(fiveyear_all_folder, 'all_area_over_time.csv'), parse_dates=['End Year'])
```

```python
# Plot multi fires over time
fig, ax = plt.subplots(figsize=(8, 8))
fire_plot.plot('End Year',
                    y='Total Area',
                    legend=False,
                    ax=ax,
                    color='red',
                    linewidth=4)
ax.set_title('Total area that experienced 2+ fires\n within 5 years in Western US Forests', size=30)
ax.set_xlabel('End year for 5 year window', size=20)
ax.set_ylabel('Area in sqkm', size=20)
plt.xticks(rotation=45, size=15)
plt.yticks(size=15)
plt.savefig(os.path.join(fiveyear_fire_folder, "fiveyear_fire.png"))
plt.show()
```

```python
# Plot multi beetles over time
fig, ax = plt.subplots(figsize=(8, 8))
beetle_plot.plot('End Year',
                    y='Total Area',
                    legend=False,
                    ax=ax,
                    color='green',
                    linewidth=4)
ax.set_title('Total area that experienced 2+ years of beetle\n kill within 5 years in Western US Forests', size=30)
ax.set_xlabel('End year for 5 year window', size=20)
ax.set_ylabel('Area in sqkm', size=20)
plt.xticks(rotation=45, size=15)
plt.yticks(size=15)
plt.savefig(os.path.join(fiveyear_beetle_folder, "fiveyear_beetle.png"))
plt.show()
```

```python
# Plot any multi disturbances over time
fig, ax = plt.subplots(figsize=(9, 9))
all_plot.plot('End Year',
                    y='Total Area',
                    legend=False,
                    ax=ax,
                    color='brown',
                    linewidth=4)
ax.set_title('Total area that experienced 2+ disturbances\n within 5 years in Western US Forests', size=30)
ax.set_xlabel('End year for 5 year window', size=20)
ax.set_ylabel('Area in sqkm', size=20)
plt.xticks(rotation=45, size=15)
plt.yticks(size=15)
plt.savefig(os.path.join(fiveyear_fire_folder, "fiveyear_all.png"), transparent=True)
plt.show()
```

```python
# Plot multi drought overtime
fig, ax = plt.subplots(figsize=(8, 8))
drought_plot.plot('End Year',
                    y='Total Area', 
                    title='Total area that experienced 2+ years of drought within 5 years in Western US Forests',
                    ylabel='Area in sqkm',
                    xlabel='Year that the 5 year window ended',
                    legend=False,
                    ax=ax)
plt.xticks(rotation=45)
plt.savefig(os.path.join(fiveyear_drought_folder, "fiveyear_drought.png"))
plt.show()
```

```python
# Plot only fire and beetle overtime
fig, ax = plt.subplots(figsize=(9, 9))
fire_beetle_plot.plot('End Year',
                    y='Total Area',
                    legend=False,
                    ax=ax,
                    color='purple',
                    linewidth=4)
ax.set_title('Total area that experienced 2+ fire and/or\ninsect/diseasedisturbances within\n5 years in Western US Forests', size=30)
ax.set_xlabel('End year for 5 year window', size=20)
ax.set_ylabel('Area in sqkm', size=20)
plt.xticks(rotation=45, size=15)
plt.yticks(size=15)
plt.savefig(os.path.join(fiveyear_fire_beetle_folder, "fiveyear_fire_beetle.png"), transparent=True, bbox_inches = 'tight')
plt.show()
```
