```python
# Import packages and define file paths
import os
import earthpy as et
from os import path
import pandas as pd
import matplotlib.pyplot as plt

home = path.join(et.io.HOME, "GitHub", "visualize-forest-disturbance")

output_data = path.join(home, "data", "output")

fire_dir = path.join(output_data, "output_tifs", "var_time_fire")
beetle_dir = path.join(output_data, "output_tifs", "var_time_beetle")
drought_dir = path.join(output_data, "output_tifs", "var_time_drought")
all_dir = path.join(output_data, "output_tifs", "var_time_all")
```

```python
# Open up fire data based on how far apart they are
threeyear_fire = pd.read_csv(path.join(fire_dir, "fire-three-year.csv"), parse_dates=['End Year'])
fouryear_fire = pd.read_csv(path.join(fire_dir, "fire-four-year.csv"), parse_dates=['End Year'])
fiveyear_fire = pd.read_csv(path.join(fire_dir, "fire-five-year.csv"), parse_dates=['End Year'])
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
threeyear_fire.plot('End Year',
                    y='Total Area',
                    ax=ax,
                    color='red',
                    label='3 year gap',
                    linewidth=4)
fouryear_fire.plot('End Year',
                    y='Total Area',
                    ax=ax,
                    color='green',
                    label='4 year gap',
                    linewidth=4)
fiveyear_fire.plot('End Year',
                    y='Total Area',
                    ax=ax,
                    color='blue',
                    label='5 year gap',
                    linewidth=4)
ax.set_title('Total area that experienced fires X years apart\nin Western US Forests', size=30)
ax.set_xlabel('End year for year window', size=20)
ax.set_ylabel('Area in sqkm', size=20)
plt.xticks(rotation=45, size=15)
plt.yticks(size=15)
plt.savefig(os.path.join(fire_dir, "fire_year_gap.png"))
plt.show()
```
